```python
from keras.models import Sequential
from keras.layers.core import Dense
from tensorflow.keras import initializers
import numpy as np
```

```python
X=np.asarray([
              [0,0],
              [0,1],
              [1,0],
              [1,1]  ])

y=np.asarray([[0],[0],[0],[1]])
```

```python
def ini_pesos(shape,dtype=None):
  return np.random.normal(loc=0.75, scale=1e-2, size= shape)
```

```python
#https://keras.io/api/layers/activations/
modelo=Sequential()
modelo.add(Dense(2, activation='sigmoid', kernel_initializer=ini_pesos,input_dim=2))
modelo.add(Dense(4, activation='sigmoid', kernel_initializer=ini_pesos))
modelo.add(Dense(1, activation='sigmoid'))

modelo.compile( loss='mean_squared_error',optimizer='adam',metrics=['accuracy'])
modelo.fit(X,y,batch_size=1,epochs=1000,verbose=0)
modelo.predict(X)
```
