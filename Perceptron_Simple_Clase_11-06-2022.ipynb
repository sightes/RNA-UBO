```python
import matplotlib.pyplot as plt 
from sklearn import datasets
import numpy as np 

#X=np.asarray(
#  [[0,0],
 #  [0,1],
##   [1,0],
 #  [1,1]])
#y=np.asarray([0,1,0,1])

X,y= datasets.make_blobs(n_samples=100,n_features=2,centers=3,cluster_std=0.24,random_state=2)

fig=plt.figure(figsize=(2,2))

plt.plot(X[:,0][y==0],X[:,1][y==0],'r^')
plt.plot(X[:,0][y==1],X[:,1][y==1],'bs')
plt.plot(X[:,0][y==2],X[:,1][y==2],'gx')
```

```python
def step_func(z):
  return 1.0 if (z>0) else 0.0 

def perceptron(X,y,a,epochs):
     m,n=X.shape
     w=np.zeros((n+1,1))
     miss=[]
     for epoch in range(epochs):
      miss_i=0
      for idx, x_i in enumerate(X):
        x_i=np.insert(x_i,0,1).reshape(-1,1)
        y_hat=step_func(np.dot(x_i.T,w))
        if (np.squeeze(y_hat)-y[idx]) >0.05 :
          w += a* ((y[idx]-y_hat)*x_i)
          miss_i +=1
      miss.append(miss_i)
     return w, miss 
 

```

```python
w,miss=perceptron(X,y,0.04,500)
```
